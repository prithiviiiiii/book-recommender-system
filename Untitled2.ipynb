```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Define the URL to scrape
url = "https://example.com/products"  # Replace with your target website

# Step 2: Send a GET request to the website
response = requests.get(url)
if response.status_code != 200:
    print("Failed to retrieve the web page")
    exit()

# Step 3: Parse the HTML content
soup = BeautifulSoup(response.content, 'html.parser')

# Step 4: Extract the data
# Assuming we are scraping product names and prices
products = []
prices = []

for item in soup.find_all("div", class_="product"):
    name = item.find("h2", class_="product-name").text.strip()
    price = item.find("span", class_="product-price").text.strip()
    products.append(name)
    prices.append(float(price.replace("$", "")))  # Remove currency symbol and convert to float

# Step 5: Create a DataFrame
d = pd.DataFrame({"Product": products, "Price": prices})

# Step 6: Perform Data Analysis
print(d.describe())  # Display basic statistics
print(d.head())  # Show the first few rows of the DataFrame

# Step 7: Visualize the Data
plt.figure(figsize=(10, 6))
plt.bar(d["Product"], d["Price"], color="skyblue")
plt.xlabel("Products")
plt.ylabel("Price")
plt.title("Product Prices")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Step 8: Save the Data
d.to_csv("products.csv", index=False)
print("Data saved to products.csv")
```
